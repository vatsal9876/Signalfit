```python
import json
import numpy as np
import faiss
```

```python
catalog=json.load(open("data/shl_catalog.json", "r", encoding="utf-8"))
```

```python
from retrieval.embeddings import generate_documents, model , generate_embeddings

path="data/embeddings.npy"

documents=generate_documents(catalog)
embeddings=generate_embeddings(documents)

np.save(path, embeddings)
```

```python
dimensions= embeddings.shape[1]
index = faiss.IndexFlatL2(dimensions)
index.add(embeddings)
faiss.write_index(index, "data/data_index")

print("Index created.")
print("Total vectors:", index.ntotal)
```
